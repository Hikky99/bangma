```python
#importing libraries. If you have already not installed these libraries, please do so first, and the import.
import numpy as np 
import pandas as pd 
import seaborn as sb
import matplotlib.pyplot as plt
from warnings import filterwarnings
filterwarnings("ignore")
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report, confusion_matrix



```

```python
#
#Reading Train_data.csv and Test_data.csv, using the pd.read_csv() function, and assig them to train and test.

#I the store test and train as variable A and B
train = pd.read_csv("Train_data.csv")
test = pd.read_csv('Test_data.csv')
A = train
B = test
```

**DataSet Checking**

```python
#.iloc[:] is used to select rows and columns of a DataFrame. Here, I specify all rows and the firt 10 columns.
A.iloc[:10]
```

```python
#.iloc[:] is used to select rows and columns of a DataFrame. Here, I specify all rows and the firt 10 columns.
B.iloc[:10]
```

```python
#I drop unnamed columns from A, the training dataset.
A = A.drop(["Unnamed: 133"],axis=1)
```

```python
#.value_counts() shows the count of each unique value (in this case, prognosis) in descending order
A.prognosis.value_counts()
```

```python
# .isna() checks for null (NaN) values. .sum() sums up values. A.isna().sum() sums up null values in A.
A.isna().sum()
```

```python
# .isna() checks for null (NaN) values. .sum() sums up values. B.isna().sum() sums up null values in B.
#noice that there no missing values in B
B.isna().sum()
```

**Define X and Y - from training data and P for testing data**

```python
Y = A[["prognosis"]]
X = A.drop(["prognosis"],axis=1)
P = B.drop(["prognosis"],axis=1)
```

**training data splitting**

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.2,random_state=42)
```

**Modeling**

```python
dtc= DecisionTreeClassifier(random_state=42)
model_dtc = dtc.fit(xtrain,ytrain)
tr_pred_dtc = model_dtc.predict(xtrain)
ts_pred_dtc = model_dtc.predict(xtest)

print("The train accuracy is:",accuracy_score(ytrain,tr_pred_dtc))
print("The test accuracy is:",accuracy_score(ytest,ts_pred_dtc))
```

```python


# Here, I Calculate additional evaluation metrics using the classification report
print("Classification Report:")
print(classification_report(ytest, ts_pred_dtc))

print("Confusion Matrix:")
print(confusion_matrix(ytest, ts_pred_dtc))
```

**Model 2 - Random Forest**

```python
rfc= RandomForestClassifier(random_state=42)
model_rfc = rfc.fit(xtrain,ytrain)
tr_pred_rfc = model_rfc.predict(xtrain)
ts_pred_rfc = model_rfc.predict(xtest)

print("training accuracy is:",accuracy_score(ytrain,tr_pred_rfc))
print("testing accuracy is:",accuracy_score(ytest,ts_pred_rfc))
```
